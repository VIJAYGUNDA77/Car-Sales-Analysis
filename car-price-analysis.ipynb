#Data visualization

---



```python
df["name"].value_counts()
```

```python
df["name"].value_counts().plot.pie(figsize=(12,8),explode=(0.1,0.01,0.01,0.01,0.01,0.01,0.01,0.01,0.01,0.01),autopct="%1.1f%%")
plt.title("Unveiling Brand Preferences",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
palette = sns.color_palette("Set2")
sns.countplot(x=df["name"], palette=palette)
plt.title("Count of Cars by Brand",fontsize=20)
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
category_counts = df["selling_price"].value_counts()
top_categories = category_counts.head(20)
plt.figure(figsize=(12, 8))
palette = sns.color_palette("Set2")
sns.countplot(x=df[df["selling_price"].isin(top_categories.index)]["selling_price"], palette=palette)
plt.xticks(rotation=45, ha="right")
plt.title("Distribution of Cars for Top 20 Selling Prices",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))

sns.displot(data=df, x="selling_price", hue="name", kind="kde")
plt.title("Selling Price Distribution Across Car Brands",fontsize=20)
```

```python
category_counts = df["km_driven"].value_counts()
top_categories = category_counts.head(20)
plt.figure(figsize=(12, 8))
palette = sns.color_palette("Set2")
sns.countplot(x=df[df["selling_price"].isin(top_categories.index)]["selling_price"], palette=palette)
plt.xticks(rotation=45, ha="right")
plt.title("Top 20 Kilometer Driven Values: Car Distribution",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))

sns.displot(data=df, x="km_driven", hue="name", kind="kde")
plt.xticks(rotation=45, ha="right")
plt.title("Kernel Density Estimate of Kilometers Driven by Car Brand",fontsize=20)
```

```python
plt.figure(figsize=(10, 6))
palette = sns.color_palette("Set2")
sns.countplot(x=df["fuel"], palette=palette)
plt.title("Distribution of Cars by Fuel Type",fontsize=20)
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
palette = sns.color_palette("Set2")
sns.countplot(x=df["seller_type"], palette=palette)
plt.title("Distribution of Cars by Seller Type",fontsize=20)
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
df["transmission"].value_counts().plot.pie(figsize=(12,8),explode=(0.1,0.01),autopct="%1.1f%%")
plt.title("Transmission Type Distribution in the Car Dataset",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(9, 5))
palette = sns.color_palette("Set2")
sns.countplot(x = df['owner'] , hue=df['name'], palette=palette)
plt.title('Distribution of Car Owners by Brand',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(9, 5))
palette = sns.color_palette("Set2")
sns.countplot(x = df['name'] , hue=df['seats'], palette=palette)
plt.title('Distribution of Car Brands by Seat Capacity',fontsize=20)
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.figure(figsize=(9, 5))
palette = sns.color_palette("Set2")
sns.countplot(x = df['name'] , hue=df['Mileage Unit'], palette=palette)
plt.title('Distribution of Car Brands by Mileage Unit',fontsize=20)
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
palette = sns.color_palette("Set2")
sns.countplot(x=df["Engine (CC)"], palette=palette)
plt.title("Engine Displacement Distribution Across Cars",fontsize=20)
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python

plt.figure(figsize=(15, 8))

sns.lineplot(x=df['year'].value_counts().index,
             y=df['year'].value_counts().values,
             alpha=1, color="r")

for i, txt in enumerate(df['year'].value_counts().index):
    plt.annotate(txt, (df['year'].value_counts().index[i],
                       df['year'].value_counts().values[i]),
                 textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9)

plt.title('Car Distribution Over the Years', fontsize=14, fontweight='heavy')
plt.show()
```

#**observations and conclusions**

---




**Based on the provided data, here are some key observations and conclusions regarding the factors that may affect the price of the cars:-**




**Brand Impact:**

Maruti and Hyundai dominate the dataset, potentially influencing resale value.

**Year Influence:**

Newer cars may command higher prices, with a concentration around 2013 and 2012.

**Price Distribution:**

Selling prices cluster around 250,000 to 550,000, with common values at 350,000 and 300,000.

**Mileage Matters:**

Lower mileage cars (10,000 to 50,000 km) may have higher resale values.

**Fuel Type Factor:**

Diesel and petrol cars prevail; diesel cars might have higher resale values.

**Seller Type Impact:**

Individual sellers dominate; individual sales may offer more competitive prices.

**Transmission Trend:**

Manual transmissions are prevalent; automatic cars may have higher resale values.

**Owner Influence:**

Fewer owners may increase resale value; first-owner cars are most common.

**Seat Count Consideration:**

Cars with more seats (e.g., 7 or 8) may have higher prices due to increased capacity.

**Power Performance:**

Cars with higher maximum power (e.g., 74.00 bhp) may have higher prices.

**Fuel Efficiency Focus:**

Most cars have kmpl mileage, emphasizing fuel efficiency.

**Engine Displacement Impact:**

Higher engine displacement may correlate with higher prices, indicating potential better performance.
